
KnapSack Problem Solution Outline
We will solve the problem using a Greedy Algorithm, followed by plotting the best solution found in each iteration to observe convergence.

Steps to Solve:
Input the data:

Items (values and weights).
Knapsack capacity.
Apply a Greedy Algorithm:

Calculate the value-to-weight ratio for each item.
Sort items in descending order of this ratio.
Add items to the knapsack while respecting the weight constraint.
Output Results:

List of selected items.
Total value and total weight.
Graphing:

Simulate a pseudo-iteration process (e.g., item-by-item selection).


```python
import matplotlib.pyplot as plt

# Input data
items = [
    {"item": 1, "value": 60, "weight": 10},
    {"item": 2, "value": 100, "weight": 20},
    {"item": 3, "value": 120, "weight": 30},
]
knapsack_capacity = 50

# Calculate value-to-weight ratio for each item
for item in items:
    item["ratio"] = item["value"] / item["weight"]

# Sort items by ratio in descending order
items.sort(key=lambda x: x["ratio"], reverse=True)

# Greedy selection process
selected_items = []
total_value = 0
total_weight = 0
iterations = []

for item in items:
    if total_weight + item["weight"] <= knapsack_capacity:
        selected_items.append(item["item"])
        total_value += item["value"]
        total_weight += item["weight"]
        iterations.append((len(selected_items), total_value))

# Output results
print("Selected Items:", selected_items)
print("Total Value:", total_value)
print("Total Weight:", total_weight)

# Plotting
x, y = zip(*iterations)
plt.plot(x, y, marker='o')
plt.title("Knapsack Problem: Value per Iteration")
plt.xlabel("Number of Items Selected")
plt.ylabel("Total Value")
plt.grid()
plt.show()
```

Explanation of the Code:

Value-to-Weight Ratio Calculation:

Each item is evaluated for its efficiency in contributing value relative to its weight.
Greedy Sorting:

Items are sorted by their efficiency.
Knapsack Filling:

Items are added one by one until the weight limit is reached.
Iteration Tracking:

Records total value at each step for plotting.
Visualization:

A line graph shows how the total value evolves with the number of items selected.

The graph will depict how the total value increases as items are added.
